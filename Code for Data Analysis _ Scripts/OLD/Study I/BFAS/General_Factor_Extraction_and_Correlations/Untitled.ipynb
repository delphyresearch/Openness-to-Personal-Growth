```python
from matplotlib import pyplot as plt
```

```python
rfr = ske.RandomForestRegressor
```

```python
df = pd.read_csv("./2021_BFAS_Study1.csv")
```

```python
opg =pd.read_csv("2021_OPG_BFAS_S1.csv")
```

```python
df.iloc[:,1:]
```

```python
regr = rfr(max_depth=2, random_state=0)
```

```python
regr.fit(df.iloc[:,1:-1],df.iloc[:,-1])
```

```python
preds = regr.predict(df.iloc[:,1:-1])
```

```python
errs = df['gfp'] - preds
```

```python
errs.abs().mean()
```

```python
errs.hist()
```

```python
fts = list(zip(regr.feature_importances_,df.iloc[:,1:].columns.values.tolist()))
```
