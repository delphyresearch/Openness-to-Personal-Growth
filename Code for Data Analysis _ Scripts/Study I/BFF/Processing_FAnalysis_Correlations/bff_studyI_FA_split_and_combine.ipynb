```python
df['absN'] = abs(df['N'])
df['absA'] = abs(df['A'])
df['absO'] = abs(df['O'])
df['absC'] = abs(df['C'])
df['absE'] = abs(df['E'])
```

```python
df = df.drop(['absN','absA','absO','absC','absE'],axis=1)
```

```python
df['maxfactor'] = df['max_factor'].apply(lambda x: x[3:])
```

```python
df = df.drop(['max_factor'],axis=1)
```

```python
col_dict={"N":2,"A":3,"C":4,"O":5,"E":6}
```

```python
def get_max_loading(x):
    mfactor = x[7]
    col = col_dict[str(mfactor)]
    return x[col]
```

```python

df['max_factor_loading'] = df.apply(lambda x:  get_max_loading(x),axis=1)
```
